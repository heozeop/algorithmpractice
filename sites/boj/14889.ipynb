```python

minVal = 1000000
visited = [0] * n
def bfs(picked, idx):
    global minVal
    global visited
    
    if picked == n // 2:
        teamVal = 0
        otherTeamVal = 0
        for i in range(n):
            for j in range(n):
                if visited[i] != visited[j]:
                    continue
                
                if visited[i]:
                    teamVal += arr[i][j]
                else:
                    otherTeamVal += arr[i][j]
        
        minVal = min(abs(otherTeamVal - teamVal), minVal)
        return
    
    for i in range(idx, n):
        if visited[i]:
            continue
        visited[i] = True
        bfs(picked + 1, i+ 1)
        visited[i] = False

bfs(0, 0)
print(minVal)
```
